```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.shape
```

```python
pd.set_option('display.max_columns', None)
```

```python
dataset
```

```python
dataset.columns
```

```python
# first we findout how many vaues have the missing values
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
#now write the information 
for feature in features_with_na:
    print(feature , np.round(dataset[feature].isnull().mean(),4) , 'missing value')
    
```

```python
features_with_na
```

```python
# now checking the how much salepricw is dependent over all the features
data = dataset.copy()
for feature in features_with_na:
    data[feature] = data[feature].isnull().astype(int)# making the missing vlaues iinto 0 or 1
    plt.figure(figsize=(4,3))
    data.groupby(feature)['SalePrice'].median().plot.bar()
   # plt.show()
  
```

```python
## numerical value
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
numerical_features # lsit contains  the numercal values
dataset[numerical_features]
```

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
dataset[year_feature]
```

```python
# now we will understand the saleprice and the year of sold
for feature in year_feature:
    plt.figure(figsize=(4,3))
    data.groupby(feature)['SalePrice'].median().plot()
    plt.xlabel(feature)
    plt.ylabel('Median House Price')
    plt.title('saleprice')
    
```

```python
# how we will compare the house age wiht hte sold prcie to understand more about at what age the price sold is high or low
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]
        plt.figure(figsize=(4,3))
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable and Discrete Variables
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
discrete_feature
```

```python
#plotting graphs for saleprices and the discrete datas
for feature in discrete_feature:
    data=dataset.copy()
    plt.figure(figsize=(4,3))
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#finding the continuous values
continuous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature+['Id']]
continuous_feature
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    plt.figure(figsize=(4,3))
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

# part 2 of eda


```python
## We will be using logarithmic transformation


for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature]:
        pass
    data[feature] = np.log(data[feature])
    plt.figure(figsize=(4,3))
    data['SalePrice'] = np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalesPrice')
    plt.title(feature)
    plt.show()
```

# outliers

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        plt.figure(figsize=(4,3))
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        
```
