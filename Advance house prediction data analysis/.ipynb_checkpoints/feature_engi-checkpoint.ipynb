```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.head()
```

```python
## Always remember there way always be a chance of data leakage so we need to split the data first and then apply feature
## Engineering
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
# first find out the categorial values and replace the nan values int othe new label 'missing'
features_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in features_nan:
        print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(), 4)))
```

```python
#replace the nan values with missing
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,features_nan)
dataset.head()
```

```python
dataset[features_nan].isnull().sum()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_na= [feature for feature in dataset if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
numerical_with_na

```

```python
for feature in numerical_with_na:
    print('{}:{} % missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in numerical_with_na:
    median_value=dataset[feature].median()
    # now we will make the new column which contains the info of feature+nan values and replace the original column with
    #the meidan values

    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature] = dataset[feature].fillna(median_value)
    
    
```

```python
dataset.head()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    dataset[feature]=dataset['YrSold']-dataset[feature]
    
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

# Numerical Variables
Since the numerical variables are skewed we will perform log normal distribution


```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].head()
```

# Handling Rare Categorical Feature
We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count() / len(dataset)
```

```python
temp
```

```python
# if the feature contains any values less than 1 % then we gonna make the new col and place it into it inside that rare col also
for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count() / len(dataset) # thiss will give me the percentage of value groupwise like
    #0.40,0.30,0.20,0.10
    #now makw the new list that contains the features whose values greater than the 1% values
    temp_df = temp[temp>0.10].index #means the temp feature will be putted into the temp_df 
    dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature],'rare_var')
```

```python
dataset.head()
```

```python

for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
