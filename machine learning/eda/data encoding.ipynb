```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple dataframe 
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df
```

```python
##create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
encoded= encoder.fit_transform(df[['color']]).toarray()
```

```python
# now make df data
df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
df
```

```python
import seaborn as sns
```

```python
df1 = sns.load_dataset('tips')
df1
```

```python
df1['sex'].unique()
```

```python
sex_encoded = encoder.fit_transform(df1[['sex']]).toarray()
sex_encoded
```

```python
sex_final = pd.DataFrame(sex_encoded,columns=encoder.get_feature_names_out())
sex_final
```

```python
# now concatinate the three columns
df_final = pd.concat([df1['sex'],sex_final])
df_final
```
