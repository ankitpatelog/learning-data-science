```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
## train test split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2, # 20 percent data into test data and remining inot train data
    random_state=42
)
```

```python
# apply linear regression 
from sklearn.linear_model import LinearRegression
# standardization of dataset
# we normally standardize the x train in regression and neutral network things 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

```

```python
# now we converted the xtrain into the standard data
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
X_test

# xtrain --> fittranform that lern form mean/standde 

# 👉 fit_transform = learn + convert
# 👉 transform = only convert (using learned values)
```

```python
# cross validation 
```

```python
# 🔹 Simple Idea (Intuition)

# Instead of training your model once on one split of data, you:
# 👉 Split your dataset multiple times
# 👉 Train and test multiple times
# 👉 Take the average performance

# This gives a more reliable result.

# 🔹 Why we use it
# Avoid overfitting (model memorizing data)
# Get a better estimate of accuracy
# Use data more efficiently (important when dataset is small


# 🔹 Example (K = 5)

# Data → split into 5 parts:

# Fold1 Fold2 Fold3 Fold4 Fold5

# Iterations:

# Train: 2,3,4,5 → Test: 1
# Train: 1,3,4,5 → Test: 2
# Train: 1,2,4,5 → Test: 3
# Train: 1,2,3,5 → Test: 4
# Train: 1,2,3,4 → Test: 5

# 👉 Final result = average of all 5 scores (give the result in mean mse)
```

```python
from sklearn.linear_model import LinearRegression
# cross validation 
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
mse = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
# 🔹 Rule of thumb

# 👉 Always compare MSE mean with your target (y) scale to determine 
```

```python
print(y.min(),y.max())
```

```python
# prediction
```

```python
reg_pred=regression.predict(X_test)
reg_pred
```
