```python
#classes in python
class students:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def display(self):
        print(self.name,self.age)
```

```python
s1 = students("Ankit",20)
s2 = students("Rashi",19)
s1.display()
s2.display()
```

```python
class car:
    def __init__(self,name,windows):
        self.name = name
        self.windows = windows

    def disp_info(self):
        print("the car name is {} and no of windows in it is {}".format(self.name,self.windows))
```

```python
car1 = car("baleno",6)
car1.disp_info()
```
